```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
auto = pd.read_csv('/kaggle/input/automobile-dataset/Automobile_data.csv')
```

```python
auto.head()
```

```python
auto.info()
```

```python
plt.figure(figsize = (6,5))
sns.barplot(data = auto, x = 'body-style', y='city-mpg')
```

```python
order = auto.groupby(['body-style']).mean().sort_values('city-mpg', ascending = False).index
sns.barplot(data = auto, x = 'body-style', y='city-mpg', order = order)
```

```python
plt.figure(figsize = (7,5))
sns.barplot(data = auto, x = 'city-mpg', y = 'body-style')
```

```python
order = auto.groupby(['body-style']).mean().sort_values('city-mpg').index
sns.barplot(data = auto , x = 'body-style', y = 'city-mpg', palette = 'GnBu', order = order)
plt.title('Body Style vs City Mileage')
plt.show()
```

```python
order = auto.groupby(['body-style']).mean().sort_values('city-mpg').index
sns.barplot(data = auto, x = 'body-style', y = 'city-mpg', hue = 'fuel-type', order = order, palette = 'BuGn')
```

```python
sns.countplot(data = auto, x = 'body-style')
```

```python
sns.countplot(data = auto, x =  'body-style', order = auto['body-style'].value_counts().index)
plt.title("No. of Categories of Cars")
```

```python
plt.figure(figsize = (10,8))
sns.scatterplot(data = auto, x = 'engine-size', y = 'wheel-base', size = 'city-mpg', sizes = (10,200), hue = 'body-style',
                palette = "bright")
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
auto['wheel_base_scaled'] = scale.fit_transform(auto[['wheel-base']])
auto['engine_size_scaled'] = scale.fit_transform(auto[['engine-size']])
```

```python
sns.scatterplot(data = auto, x = 'wheel_base_scaled', y = 'engine_size_scaled', marker = "o", s = 40, hue = 'body-style')
```

```python
# Lets see box plot of horsepower wrt fuel type
```

```python
auto.replace("?", 0, inplace = True)
```

```python
auto['horsepower'].replace("?",0, inplace = True)
auto['horsepower'] = auto['horsepower'].astype('int')
```

```python
sns.boxplot(data = auto , x = 'fuel-type', y = 'horsepower')
```

```python
auto['price'] = auto['price'].astype('int')
auto['price'] = auto['price'].fillna(auto['price'].mean())
```

```python
sns.boxplot(data = auto, x = 'body-style', y = 'price' )
```
